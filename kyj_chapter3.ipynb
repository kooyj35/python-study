###03-1 불 자료형과 if 조건문

```python
#--Boolean, 불린, 불리언 ; 오직 true와 false값만 가질 수 있다

print(True)
print(False)

##불 만들기 : 비교 연산자
print(10 == 100)
print(10 != 100)
print(10 < 100)
print(10 > 100)
print(10 <= 100)
print(10 >= 100)


print("가방" == "가방")
print("가방" != "하마")
print("가방" < "하마")
print("가방" > "하마")

x=25
print(10 < x < 30)
print(40 < x < 60)
```

```python
##불 연산하기: 논리 연산자
#--not ; 아니다; 불을 반대로 전환
#--and; 그리고; 피연산자 두개가 모두 참일 때 true 출력, 그 외 모두 false
#--or; 또는; 피연산자 두개 중에 하나만 참이어도 true, enro 모두 거짓일때만 false

#NOT 연산자; 단항 연산자로(피연산자 1개) 참과 거짓을 반대로 바꿀 때 사용
print(not True)
print(not False)

#손코딩_not 연산자 조합_159p
x = 10
under_20 = x < 20
print("under_20:", under_20)
print("not under_20:", not under_20)


#AND연산자(양쪽 변의 값이 모두 참일 때만)와 OR연산자(둘 중 하나만 참이어도 true)
print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

```python
# IF 조건문이란?
#--파이썬에서 if 조건문은 조건에 따라 코드를 실행하거나, 실행하지 않게 만들고 싶을 때
#--코드의 실행 흐름을 변경한다는 뜻, 이렇게 조건을 기반으로 실행의 흐름을 변경하는 것을 '조건분기'

if True:
        print("True입니다...!")
        print("정말 True입니다...!")
        
```

```python
#if 뒤에 있는 불 값이 거짓인 경우에는 들여쓰기 된 명령문이 있어도 실행되지 않음
####들여쓰기는 코드 앞에 '띄어쓰기 4번', '띄어쓰기 2번','탭'과 같음

if False:
        print("False입니다...!")

        
```

```python
##오전과 오후를 구분하는 프로그램
#날짜/시간과 관련된 기능을 가져오기
import datetime

#현재 날짜/시간 구한다
now = datetime.datetime.now()

#오전 구분
if now.hour < 12:
    print("현재 시각은 {}시로 오전입니다!".format(now.hour))

#오후 구분
if now.hour > 12:
    print("현재 시각은 {}시로 오후입니다!".format(now.hour))
```

```python
##계절을 구분하는 프로그램
#날짜/시간과 관련된 기능을 가져오기
import datetime

#현재 날짜/시간 구한다
now = datetime.datetime.now()

#봄 구분
if 3 <= now.month <= 5:
    print("이번 달은 {}월로 봄입니다!".format(now.month))

#여름 구분
if 6 <= now.month <= 8:
    print("이번 달은 {}월로 여름입니다!".format(now.month))

#가을 구분
if 9 <= now.month <= 11:
    print("이번 달은 {}월로 가을입니다!".format(now.month))

#겨울 구분
if  now.month == 12 or 1 <= now.month <= 2:
    print("이번 달은 {}월로 겨울입니다!".format(now.month))
```

```python
##짝수와 홀수 구분하기 p169

#입력 받기
number = input("정수 입력>")
#마지막 자리 숫자 추출
last_character = number[-1]
#숫자로 변환
last_number = int(last_character)
#짝수 확인
if last_number == 0\
    or last_number == 0\
    or last_number == 2\
    or last_number == 4\
    or last_number == 6\
    or last_number == 8:
    print("짝수입니다")

#홀수 확인
if last_number == 1\
    or last_number == 3\
    or last_number == 5\
    or last_number == 7\
    or last_number == 9:
    print("홀수입니다")

#---------------------------------------------------------------------------------------
###직접 손코딩 1__in문자열 연산자를 활용하여 짝수와 홀수 구분
number = input("정수 입력>")
last_character = number[-1]
if last_character in "02468":
    print("짝수다")
if last_character in "13579":
    print("홀수다")


#---------------------------------------------------------------------------------------
###직접 손코딩 2__나머지 연산자를 활용하여 짝수와 홀수 구분
number = input("정수 입력>")
number = int(number)
if number %2 == 0:
    print("짝수")
if number %2 == 1:
    print("홀수")

```

###03-2 if~else 와 elif 구문 p176

```python
#else 조건문의 활용
#정반대되는 상황에서(홀수,짝수) 두 번이나 조건을 비교하는 것은 낭비
#else 구문은 if 조건문 뒤에 사용하며 if 조건문의 조건이 거짓일 때 실행

#if 조건문에 else 구문을 추가해서 짝수와 홀수 구분
number = input("정수 입력>")
number = int(number)
if number %2 == 0:
    #조건이 참일 때, 즉 짝수 조건
    print("짝수입니다")
else:
    #조건이 거짓일 때, 즉 홀수 조건
    print("홀수입니다")
```

```python
##조건문이 오로지 두 가지로만 구분 될 때 if else 구문을 사용하면 조건 비교를 한 번만 함


#elif 구문
#두가지로만 구분되지 않을 때, 즉 계절, 요일 등 세 개 이상의 조건을 연결해서 사용
#elif 구문은 if 조건문과 else 구문 사이에 입력

###계절 구하기
    #날짜/시간과 관련된 기능을 가져온다
import datetime
    #현재 날짜/시간을 구하고
    #쉽게 사용할 수 있게 월을 변수에 저장한다
now = datetime.datetime.now()
month = now.month
    #조건문으로 계절을 확인한다
if 3 <= month <= 5:
    print("현재는 봄입니다.")
elif 6 <= month <= 8:
    print("현재는 여름입니다.")
elif 9<= month <= 11:
    print("현재는 가을입니다.")
else:
    print("현재는 겨울입니다.")
```

```python
## if 조건문을 효율적으로 사용하기

#변수 선언
score = float(input("학점 입력>"))

#조건문을 적용
if score == 4.5:
    print("신")
elif 4.2 <= score < 4.5:
    print("교수님의 사랑")
elif 3.5 <= score < 4.2:
    print("현 체제의 수호자")
elif 2.8 <= score < 3.5:
    print("일반인")
elif 2.3 <= score < 2.8:
    print("일탈을 꿈꾸는 소시민")
elif 1.75 <= score < 2.3:
    print("오란문화의 선구자")
elif 1.0 <= score < 1.75:
    print("불가촉천민")
elif 0.5 <= score < 1.0:
    print("자벌레")
elif 0 <= score < 0.5:
    print("플랑크톤")
elif score == 0:
    print("시대를 앞서가는 혁명의 씨앗")



###효율적으로 재구현
###if 조건문은 위에서 아래로 흐르며, else 구문과 elif구문은 이전의 조건이 맞지 않을 때
##넘어오는 부분이므로 위에서 이미 제외된 조건을 한버 더 검사할 필요가 없다
#변수 선언
score = float(input("학점입력>"))

#조건문을 적용
if score == 4.5:
    print("신")
elif 4.2 <= score:
    print("교수님의 사랑")
elif 3.5 <= score:
    print("현 체제의 수호자")
elif 2.8 <= score:
    print("일반인")
elif 2.3 <= score:
    print("일탈을 꿈꾸는 소시민")
elif 1.75 <= score:
    print("오란문화의 선구자")
elif 1.0 <= score:
    print("불가촉천민")
elif 0.5 <= score:
    print("자벌레")
elif 0 <= score:
    print("플랑크톤")
else:
    print("시대를 앞서가는 혁명의 씨앗")
```
