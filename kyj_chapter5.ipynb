```python
# 함수의 기본
# def 함수 이름():
    # 문장

#손코딩_p275

def print_3_times():
    print("안녕하세요")
    print("안녕하세요")
    print("안녕하세요")

print_3_times()
```

```python
# 함수에 매개변수 만들기
# 매개변수는 함수를 생성할 때 괄호 내부에 식별자를 입력해서 만든다
# def 함수 이름(매개변수, 매개변수, ...):
    # 문장

#손코딩_p276


#함수 괄호 내부에 value,n 이라는 식별자 입력, 이는 매개변수가된다. 즉
#value에는 "안녕하세요"가 들어가고 매개변수 n에는 5가 들어간다
def print_n_times(value, n):
    for i in range(n):
        print(value)

print_n_times("안녕하세요", 5)
```

```python
#매개변수와 관련된 typeError
#함수를 호출할 때 매개변수를 넣지 않거나, 더 많이 넣으면?

def print_n_times(value, n):
    for i in range(n):
        print(value)

print_n_times("안녕하세요")
```

```python
#매개변수와 관련된 typeError
#함수를 호출할 때 매개변수를 넣지 않거나, 더 많이 넣으면?
#지정한 개수보다 더 많이 넣었을 경우

def print_n_times(value, n):
    for i in range(n):
        print(value)

print_n_times("안녕하세요",10,20)
```

```python
#가변 매개변수
#--함수를 선언할 때 매개변수와 함수를 호출할 때의 매개변수가 같아야 하낟.
#--매개변수를 원하는 만큼 받을 수 있는 함수를 '가변 매개변수'라 하고 매개변수 개수가 변할 수 있다.

# def 함수이름(매개변수, 매개변수, ..., *가변 매개변수):
    # 문장

## 가변 매개변수 사용시 제약
#1. 가변 매개변수 뒤에는 일반 매개변수가 올 수없다.
#2. 가변 매개변수는 하나만 사용할 수 있습니다.


def print_n_times(n, *values):
    #n번 반복
    for i in range(n):
        #values는 리스트처럼 활용
        for value in values:
            print(value)
            #단순 줄바꿈
            #print()

# 함수를 호출
print_n_times(3, "안녕하세요", "즐거운", "파아썬 프로그래밍")



### 그리고 가변 매개변수 뒤에는 일반 매개변수가 올수 없다고 했는데, 즉 만약
# print_n_times("안녕하세요", "즐거운", "파아썬 프로그래밍", 3) 처엄 사용한다면 어디까지가 가변매개변수이고
# 어디까지가 n인지 구분하기 힘들다. 
```

```python
## 기본 매개변수

def print_n_times(value, n=2):   #<---매개변수 n을 n=2라는 형태로 입력
    #n번 반복
    for i in range(n):
        print(value)

    #함수 호출
print_n_times("안녕")
```

```python
## 키워드 매개변수
# 기본 매개변수가 가변 매개변수보다 앞에 올 때

def print_n_times(n=2, *values):
    #n번 반복
    for i in range(n):
        #values는 리스트처럼 활용
        for value in values:
            print(value)

print_n_times("안녕", "즐거운", "python")
```

```python
## 키워드 매개변수
# '가변' 매개변수가 기본 매개변수보다 앞에 올 때

def print_n_times(*values, n=2):
    #n번 반복
    for i in range(n):
        #values는 리스트처럼 활용
        for value in values:
            print(value)
        #단순 줄바꿈
        print()

print_n_times("안녕", "즐거운", "python", 3)
```
