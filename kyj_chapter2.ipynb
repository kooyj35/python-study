02-2 숫자   

정수integer; 소수점이 없는 숫자     
실수floating point(부동 소수점_소수점이 움직이는 숫자); 소수점이 있는 숫자

이 때 0은 소수점이 없는 숫자이고 0.0은 소수점이 있는 숫자   
같은 크기를 나타낸다 해도 소수점이 있느냐 없느냐에 따라 숫자의 자료형이 달라짐

```python
##숫자의 종류

print(273)
print(52.273)

print(type(52))
print(type(52.273))


print(0)
print(type(0))
print(0.0)
print(type(0.0))
#프로그래밍 언어에서는 두 자료형을 구분해서 사용
#문자열 선택 연산자([])등을 사용할 때 괄호 안 숫자가 정수가 아니라 부동 소수점이면 오류가 발생
```

```python
## 숫자 연산자
#사칙연산자
print("5+7=",5+7)
print("5-7=",5-7)
print("5*7=",5*7)
print("5/7=",5/7)
    
#정수 나누기 연산자://
print("3/2=",3/2)
print("3//2=",3//2) 

#나머지 연산자:%
print("5 % 2=",5%2)

#제곱 연산자:**
print("2 ** 1=", 2**1)
print("2 ** 2=", 2**2)
print("2 ** 3=", 2**3)
print("2 ** 4=", 2**4)
```

```python
#02-4 숫자와 문자열의 다양한 기능    

#format함수; 중괄호{}를 포함한 문자열 뒤에 마침표를 찍고 format 함수 사용하는데
#중괄호의 개수와 format 함수 괄호 안 매개변수의 개수는 반드시 같아야 한다.

"{}".format(10) 
"{}{}".format(10,20)    
"{}{}{}{}{}".format(101,102,103,104,105)
```

```python
#손코딩_p135
#format()함수로 숫자를 문자열로 변환
string_a = "{}".format(10)

#출력
print(string_a)
print(type(string_a))


#format 함수로 숫자를 문자열로 변환
format_a = "{}만 원".format(5000)
format_b = "파이썬 열공하여 첫 연봉 {}만들기".format(5000)
format_c = "{} {} {}".format(3000,4000,5000)
format_d = "{} {} {}".format(1, "문자열", True)

#출력
print(format_a)
print(format_b)
print(format_c)
print(format_d)
```

```python
#format 함수의 다양한 기능
#손코딩_p137

#정수
output_a = "{:d}".format(52)

#특정 칸에 출력
output_b = "{:5d}".format(52)
output_c = "{:10d}".format(52)

#빈칸 0으로 채움
output_d = "{:05d}".format(52)
output_e = "{:05d}".format(-52)

print('# 기본')
print(output_a)
print('# 특정 칸에 출력하기')
print(output_b)
print(output_c)
print('# 빈칸을 0으로 채우기')
print(output_d)
print(output_e)

#------------------------------------------------------------------------------------
#조합하기
output_h = "{:+5d}".format(52)  #기호를 뒤로 밀기: 양수
output_i = "{:+5d}".format(-52) #기호를 뒤로 밀기: 양수
output_j = "{:=+5d}".format(52) # 기호를 앞으로 밀기: 양수
output_k = "{:=+5d}".format(-52) #기호를 앞으로 밀기: 음수
output_l = "{:+05d}".format(52) #0으로 채우기 :양수
output_m = "{:+05d}".format(-52) #0으로 채우기: 음수
##기호와 공백을 조합할 때는 = 기호를 앞에 붙일 수 있다.
#예를 들어 5칸의 공간을 잡았을 때 기호를 빈칸 앞에 붙일지, 숫자 앞에 붙일지 지정


print("# 조합하기")
print(output_h)
print(output_i)
print(output_j)
print(output_k)
print(output_l)
print(output_m)
```

```python
#부동 소수점 출력의 다양한 형태
#float 자료형 출력을 강제로 지정할 때는 {:f} 사용


#다양한 형태의 부동 소수점 출력
output_a = "{:f}".format(52.273)
output_b = "{:15f}".format(52.273)
output_c = "{:+15f}".format(52.273)
output_d = "{:+015f}".format(52.273)

print(output_a)
print(output_b)
print(output_c)
print(output_d)


#소수점 아래 자릿수 지정
##15칸을 잡고 소수점을 각각 3자리,2자리, 1자리로 출력하고 이때 자동으로 반올림도 일어남
output_a = "{:15.3f}".format(52.273)
output_b = "{:15.2f}".format(52.273)
output_c = "{:15.1f}".format(52.273)

print(output_a)
print(output_b)
print(output_c)


#의미 없는 소수점 제거하기
###파이썬은 0과 0.0을 출력했을 때 내부적으로 자료형이 다르므로, 서로 다른 값을 출력
###의미 없는 0을 제거한 후 출력하고 싶을 때 {:g} 사용
output_a = 52.0
output_b = "{:g}".format(output_a)
print(output_a)
print(output_b)


#대소문자 바꾸기 upper, lower
a = "Hello Python Programming..!"
a.upper()
a.lower()


#문자열 양옆의 공백 제거 strip()
##strip(); 양옆 공백 제거
##lstrip(); 왼쪽 공백 제거
##rstrip(); 오른쪽 공백 제거

input_a = """
        안녕하세요
문자열의 함수를 알아봅니다.
        """
print(input_a)
print(input_a.strip())

#문자열의 구성 파악하기 is~()
##isalnum 알파벳 또는 숫자로만 구성되어 있는지 
##isalpha 알파벳으로만 구성되어 있는지
##isdentifier 식별자로 사용할 수 있는지
##isdecimal 정수 형태인지
##isdigit 숫자로 인식될 수 있는지
##isspace 공백으로만 구성되어 있는지
##islower 소문자로만 
##isupper 대문자로만

print("TrainA10".isalnum())
print("10".isdigit())


#문자열 찾기 find, rfind
##find() ; 왼쪽부터 찾아서 처음 등장하는 위치
##rfind() ; 오른쪽부터 찾아서 

output_a = "안녕안녕하세요".find("안녕")
print(output_a)
output_b = "안녕안녕하세요".rfind("안녕")
print(output_b)


#문자열과 in 연산자
##문자열 내부에 어떤 문자열이 있는지 확인할 때, 출력은 True or False
print("안녕" in "안녕하세요")
print("잘자" in "안녕하세요")

#문자열 자르기: split()
##문자열을 특정한 문자로 자를 때
a = "10 20 30 40 50".split(" ") #split 함수 괄호 안의 공백을 기준으로 자른다.
print(a)
```

```python
#f-문자열

##--str()은 표현식을 문자열로 바꿔주어 TypeError오류를 피하기 위해 사용
##--format()은 매개변수를 {}기호 위치에 대치해 자료형 제약없이 문자열에 데이터 삽입
## f-문자열은 파이썬 3.6버전부터 format()함수를 더 간단하게 사용
## f'문자열{표현식}문자열'

"{}".format(10)
f'{10}'

f"3+4={3+4}"
f"""1+2={1+2}
    2+3={2+3}
    3+4={3+4}"""
```

```python
# f-문자열보다 format()함수를 사용하는 것이 더 좋은 경우

#1. 문자열 내용이 너무 많을 때
name = "구름"
age = 7
"""#문자열이 너무 긴 경우
문자열이 너무 긴 상황에서
데이터 {}를 출력해야 하는 경우가 있다.

이때 f-문자열을 사용하면
어떤 위치에 어떤 데이터{}가 출력되는지 확인하기 위해서
문자열 전체를 읽어야 하는 문제가 있다.
이러한 경우 format()함수를 사용하는 것이 편리하다.
문자열이 아무리 길어도 다음 줄만 보면
어떤 데이터를 출력하는지 쉽게 알 수 있다.
""".format(name, age)
```
