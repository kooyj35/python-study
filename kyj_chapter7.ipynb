###### Query String Parameters ÎòêÎäî Request Payload ÏÑπÏÖòÏùÑ ÌôïÏù∏ÌïòÏó¨, Ïù¥ ÏöîÏ≤≠ÏùÑ ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Î≥¥ÎÇ¥Í∏∞ ÏúÑÌï¥ ÌïÑÏöîÌïú **Îß§Í∞úÎ≥ÄÏàò(Parameters, Ïòà: ÏßÄÏó≠ ÏΩîÎìú, ÎÇ†Ïßú Îì±)**Í∞Ä Î¨¥ÏóáÏù∏ÏßÄ ÌååÏïÖÌï©ÎãàÎã§.

```python
import requests
from bs4 import BeautifulSoup
import sys

# Network ÌÉ≠ÏóêÏÑú Ï∞æÏùÄ URL
TARGET_URL = "https://www.weather.go.kr/w/wnuri-eng/weather/mid-term-forecast.do?stnId1=108&stnId2="

def extract_simple_table_data():
    """Ï§ëÍ∏∞ ÏòàÎ≥¥ HTML ÌÖåÏù¥Î∏îÏùò Îç∞Ïù¥ÌÑ∞Î•º Í∞ÑÍ≤∞ÌïòÍ≤å Ï∂îÏ∂úÌïòÏó¨ Ï∂úÎ†•Ìï©ÎãàÎã§."""
    try:
        # 1. URL ÏöîÏ≤≠ Î∞è ÏùëÎãµ ÌôïÏù∏
        response = requests.get(TARGET_URL)
        response.raise_for_status() # HTTP ÏÉÅÌÉú ÏΩîÎìúÍ∞Ä 400 Ïù¥ÏÉÅÏù¥Î©¥ ÏòàÏô∏ Î∞úÏÉù

        # 2. BeautifulSoupÏúºÎ°ú HTML ÌååÏã±
        soup = BeautifulSoup(response.text, "html.parser")

        # 3. Î™©Ìëú ÌÖåÏù¥Î∏î Ï∞æÍ∏∞ (ÌÅ¥ÎûòÏä§ Ïù¥Î¶ÑÏúºÎ°ú Ï∞æÍ∏∞)
        # Ïù¥ ÌÖåÏù¥Î∏î ÏïàÏóê ÏòàÎ≥¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Îì§Ïñ¥ ÏûàÏäµÎãàÎã§.
        forecast_table = soup.find('table', class_='table-col table-zebra')

        if not forecast_table:
            print("Ïò§Î•ò: Ï§ëÍ∏∞ ÏòàÎ≥¥ ÌÖåÏù¥Î∏îÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")
            sys.exit(1)

        # 4. ÌÖåÏù¥Î∏î ÎÇ¥Ïö© Ï∂îÏ∂ú Î∞è Ï∂úÎ†•
        print("--- Ï§ëÍ∏∞ ÏòàÎ≥¥ ÌÖåÏù¥Î∏î Îç∞Ïù¥ÌÑ∞ (Í∞ÑÍ≤∞ Ï∂îÏ∂ú) ---")
        
        # ÌÖåÏù¥Î∏îÏùò Î™®Îì† Ìñâ(tr)ÏùÑ Ï∞æÏäµÎãàÎã§.
        rows = forecast_table.find_all('tr')

        for i, row in enumerate(rows):
            # ÌñâÏùò Î™®Îì† ÏÖÄ(th ÎòêÎäî td)ÏùÑ Ï∞æÏäµÎãàÎã§.
            cols = row.find_all(['th', 'td'])
            
            # ÏÖÄÏùò ÌÖçÏä§Ìä∏Îßå Ï∂îÏ∂úÌïòÏó¨ Î¶¨Ïä§Ìä∏Î°ú ÎßåÎì≠ÎãàÎã§. (Í≥µÎ∞± Ï†úÍ±∞)
            row_data = [col.text.strip() for col in cols if col.text.strip()]
            
            if row_data:
                # Ìñâ Î≤àÌò∏ÏôÄ Ï∂îÏ∂úÎêú Îç∞Ïù¥ÌÑ∞Î•º Ï∂úÎ†•Ìï©ÎãàÎã§.
                # Ïù¥ Îç∞Ïù¥ÌÑ∞Î•º Î≥¥Í≥† Ïñ¥Îäê Ïó¥Ïóê 'ÎÇ†Ïî®'ÏôÄ 'Í∏∞Ïò®'Ïù¥ ÏûàÎäîÏßÄ ÌååÏïÖÌï¥Ïïº Ìï©ÎãàÎã§.
                print(f"[{i:02} Ìñâ]: {row_data}")
        
        print("------------------------------------------")

    except requests.exceptions.RequestException as e:
        print(f"URL ÏöîÏ≤≠ Ï§ë Ïò§Î•ò Î∞úÏÉù: {e}")
        sys.exit(1)
    
    except Exception as e:
        print(f"Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨ Ï§ë ÏòàÏÉÅÏπò Î™ªÌïú Ïò§Î•ò Î∞úÏÉù: {e}")
        sys.exit(1)

if __name__ == "__main__":
    extract_simple_table_data()
```
