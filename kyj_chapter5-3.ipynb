```python
#í•¨ìˆ˜ì„ ì–¸
def power(item):
    return item*item
def under_3(item):
    return item < 3

#ë³€ìˆ˜ì„ ì–¸
list_input_a = [1,2,3,4,5]


# map()
output_a = map(power, list_input_a)
print("# map()í•¨ìˆ˜ì˜ ì‹¤í–‰ ê²°ê³¼")
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()


# filter()
output_b = filter(under_3, list_input_a)
print("# filter()í•¨ìˆ˜ì˜ ì‹¤í–‰ ê²°ê³¼")
print("filter(under_3, list_input_a):", output_b)
print("filter(under_3, list_input_a):", list(output_b))
print()
```

```python
# info.txtì— ìƒì„±ëœ ë°ì´í„°
import random
hanguls = list("ê°€ë‚˜ë‹¤ë¼ë§ˆë°”ì‚¬ì•„ìì°¨ì¹´íƒ€íŒŒí•˜")
with open ("info.txt","w") as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40, 100)
        height = random.randrange(140, 200)
        file.write("{},{},{}\n".format(name,weight,height))
        
```

```python
# ë°ì´í„° í•œì¤„ì”© ì½ê¸° for ë°˜ë³µë¬¸

#ğŸ’ for í•œ ì¤„ì„ ë‚˜íƒ€ë‚´ëŠ” ë¬¸ìì—´ in íŒŒì¼ ê°ì²´:


with open("info.txt", "r") as file:
    for line in file:
        #ë³€ìˆ˜ì„ ì–¸
        (name, weight, height) = line.strip().split(",")

        if(not name) or (not weight) or (not height):
            continue
        bmi = int(weight)/((int(height)/100)**2)
        result = ""
        if 25 <= bmi:
            result = "ê³¼ì²´ì¤‘"
        elif 18.5 <= bmi:
            result = "ì •ìƒ ì²´ì¤‘"
        else:
            result = "ì €ì²´ì¤‘"

print('\n'.join([
    "ì´ë¦„: {}",
    "ëª¸ë¬´ê²Œ: {}",
    "í‚¤: {}",
    "BMI: {}",
    "ê²°ê³¼: {}",
]).format(name,weight,height,bmi,result))
print()
```

```python
# ì œë„ˆë ˆì´í„° í•¨ìˆ˜
def test():
    print("í•¨ìˆ˜ê°€ í˜¸ì¶œë˜ì—ˆìŒ")
    yield "test"

# í•¨ìˆ˜í˜¸ì¶œ
print("A ì§€ì  í†µê³¼")
test()

print("A ì§€ì  í†µê³¼")
test()
print(test())
```

```python
#ì œë„ˆë ˆì´í„° ê°ì²´ì™€ next()í•¨ìˆ˜

def test():
    print("A ì§€ì  í†µê³¼")
    yield 1
    print("B ì§€ì  í†µê³¼")
    yield 2
    print("C ì§€ì  í†µê³¼")

output = test()

print("D ì§€ì  í†µê³¼")
a = next(output)
print(a)
print("E ì§€ì  í†µê³¼")
b = next(output)
print(b)
print("F ì§€ì  í†µê³¼")
c = next(output)
print(c)

next(output)
```
